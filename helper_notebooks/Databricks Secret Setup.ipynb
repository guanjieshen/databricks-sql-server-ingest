## Databricks Secret Setup
This notebook helps you create and manage Databricks secrets using the Databricks CLI.

```python
# List all secret scopes
dbutils.secrets.listScopes()
```

```python
# Store a secret in the scope
scope_name = "sql-ingestion"  # Your scope name
key_name = "guanjiesqldb"       # Your secret key name

print("To create the secret scope, run:")
print(f"databricks secrets create-scope {scope_name}\n")
print("To store a secret, run:")
print(f"databricks secrets put-secret {scope_name} {key_name}")
print("\nYou'll be prompted to enter the secret value in an editor.")
```

```python
# Check if scope exists first
existing_scopes = [scope.name for scope in dbutils.secrets.listScopes()]

if scope_name in existing_scopes:
    try:
        secrets = dbutils.secrets.list(scope_name)
        print(f"✓ Secrets in '{scope_name}':")
        for secret in secrets:
            print(f"  - {secret.key}")
        
        print(f"\nUse: dbutils.secrets.get(scope='{scope_name}', key='{key_name}') to retrieve")
        print(f'\nUse: "{{{{secrets/{scope_name}/{key_name}}}}}" to retrieve')
    except Exception as e:
        print(f"Error: {e}")
else:
    print(f"⚠ Scope '{scope_name}' does not exist yet.")
    print(f"\nCreate it by running this command in your terminal:")
    print(f"  databricks secrets create-scope {scope_name}")
    print(f"\nThen store your secret with:")
    print(f"  databricks secrets put-secret {scope_name} {key_name}")
    print(f"\nAfter creating the scope, re-run this cell to verify.")
```
